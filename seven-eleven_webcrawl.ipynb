## 서울 중구에 대한 것만 가져와보고 패턴 파악

```python
from selenium import webdriver
driver = webdriver.Chrome()


url = "https://www.7-eleven.co.kr/"
driver.get(url)
```

```python
from selenium.webdriver.common.by import By
target_store = "#header > div > div > div.head_util > a.util_store.store_open"
driver.find_element(By.CSS_SELECTOR, target_store).click()
```

```python
from bs4 import BeautifulSoup as BS
seven = BS(driver.page_source)
seven
```

```python
import re
seven_total = []
for tmp in seven.find("div", class_="list_stroe").findAll("li"):
    seven_dict = {}
    seven_dict["offeringService"] = [x["alt"] for x in tmp.findAll("img")]
    seven_dict["shopname"] = tmp.find("span").text.strip()
    seven_dict["address"] = ' '.join(tmp.findAll("span")[-2].text.split())
    if len(seven_dict["address"]) < 2:
        ' '.join(tmp.findAll("span")[-3].text.split())
    
    _, lat, lon = re.findall("(?<=\().+(?=\))", tmp.find('a')['href'])[0].split(",")
    seven_dict["latitude"] = lat
    seven_dict["longitude"] = lon
    #print(re.findall("(?<=\().+(?=\))", tmp.find('a')['href'])[0].split(","))    
    
   
    seven_total.append(seven_dict)
seven_total
```

```python
len(seven_total)
```

```python
import pandas as pd
pd.DataFrame(seven_total)
```

# 전국 세븐일레븐 웹크롤링

```python
import time

seven_total = []

for x in range(2,19):
    driver.find_element(By.CSS_SELECTOR, f"#storeLaySido > option:nth-child({x})").click()
    time.sleep(2)
    for y in range(2,50):
        try:
            driver.find_element(By.CSS_SELECTOR,  "#storeLayGu > option:nth-child({})".format(y)).click()
            driver.find_element(By.CSS_SELECTOR, "#storeButton1").click()
            seven = BS(driver.page_source)
            
            for tmp in seven.find("div", class_="list_stroe").findAll("li"):
                seven_dict = {}
                seven_dict["offeringService"] = [x["alt"] for x in tmp.findAll("img")]
                seven_dict["shopname"] = tmp.find("span").text.strip()
                seven_dict["address"] = ' '.join(tmp.findAll("span")[-2].text.split())
                if len(seven_dict["address"]) < 2:
                    ' '.join(tmp.findAll("span")[-3].text.split())

                _, lat, lon = re.findall("(?<=\().+(?=\))", tmp.find('a')['href'])[0].split(",")
                seven_dict["latitude"] = lat
                seven_dict["longitude"] = lon
                seven_total.append(seven_dict)
            time.sleep(2)
        except:
            break
```
